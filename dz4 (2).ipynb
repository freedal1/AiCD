```python
# 11
import random
import statistics

def test_sorting_algorithm(sort_func, f_gen, repetitions, list_lengths):
    results = []
    
    for length in list_lengths:
        comparisons = []
        swaps = []
        
        for _ in range(repetitions):
            lst = f_gen(length)
            lst_copy = lst.copy()
            
            sort_func(lst)
            
            comparisons.append(sort_func.comparisons)
            swaps.append(sort_func.swaps)
            
            # Reset comparison and swap counters for the next iteration
            sort_func.comparisons = 0
            sort_func.swaps = 0
        
        results.append({
            'length': length,
            'min_comparisons': min(comparisons),
            'avg_comparisons': statistics.mean(comparisons),
            'max_comparisons': max(comparisons),
            'min_swaps': min(swaps),
            'avg_swaps': statistics.mean(swaps),
            'max_swaps': max(swaps)
        })
    
    return results

# Пример использования функции test_sorting_algorithm:
def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                bubble_sort.swaps += 1
            bubble_sort.comparisons += 1

bubble_sort.comparisons = 0
bubble_sort.swaps = 0

def generate_random_list(length):
    return random.sample(range(length * 10), length)

list_lengths = [10, 100, 1000]
repetitions = 10

results = test_sorting_algorithm(bubble_sort, generate_random_list, repetitions, list_lengths)

for result in results:
    print(result)

```

```python
# 12
def cocktail_shaker_sort(arr):
    n = len(arr)
    comparisons = 0 # считает кол-во сравнений
    swaps = 0 # считает кол-во перестановок
    start = 0 # начало секции
    end = n - 1 # конец секции
    while start <= end: #Пока start не достигнет end, продолжаем итерации, т.к. раздел ещё не отсортирован.
        swapped = False
        
        # Проход слева направо (поднятие максимума)
        for i in range(start, end):
            comparisons += 1
            if arr[i] > arr[i+1]: #Если текущий больше следующего, меняем их местами
                arr[i], arr[i+1] = arr[i+1], arr[i]
                swaps += 1
                swapped = True
        end -= 1
        
        # Проход справа налево (понижение минимума)
        for i in range(end, start, -1):
            comparisons += 1
            if arr[i] < arr[i-1]: #Если текущий меньше следующего, меняем их местами
                arr[i], arr[i-1] = arr[i-1], arr[i]
                swaps += 1
                swapped = True
        start += 1
        
        if not swapped:
            break
    
    return arr, comparisons, swaps

# Пример списка для сортировки
arr = [3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]
print("Исходный список:", arr)

# Сортировка списка с помощью "камешек-пузырька"
sorted_arr, comparisons, swaps = cocktail_shaker_sort(arr.copy())

print("Отсортированный список:", sorted_arr)
print("Количество сравнений:", comparisons)
print("Количество перестановок:", swaps)
```

```python
# 14
import random
import statistics

# Алгоритм сортировки пузырьком
def bubble_sort(arr):
    n = len(arr)
    comparisions = 0
    swaps = 0
    for i in range(n):
        for j in range(0, n-i-1):
            comparisions += 1
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swaps += 1
    return arr, comparisions, swaps

# Алгоритм "камешек-пузырька"
def cocktail_shaker_sort(arr):
    n = len(arr)
    comparisons = 0
    swaps = 0
    start = 0
    end = n - 1
    while start <= end:
        swapped = False
        for i in range(start, end):
            comparisons += 1
            if arr[i] > arr[i+1]:
                arr[i], arr[i+1] = arr[i+1], arr[i]
                swaps += 1
                swapped = True
        end -= 1
        for i in range(end, start, -1):
            comparisons += 1
            if arr[i] < arr[i-1]:
                arr[i], arr[i-1] = arr[i-1], arr[i]
                swaps += 1
                swapped = True
        start += 1
        if not swapped:
            break
    return arr, comparisons, swaps

# Генерация отсортированного списка
def generate_sorted_list(length):
    return list(range(1, length + 1))

# Генерация отсортированного в обратном порядке списка
def generate_reverse_sorted_list(length):
    return list(range(length, 0, -1))

# Генерация почти отсортированного списка
def generate_almost_sorted_list(length):
    lst = list(range(1, length + 1))
    random_index = random.randint(1, length-1)
    lst[random_index], lst[random_index - 1] = lst[random_index - 1], lst[random_index]
    return lst

# Генерация случайного списка
def generate_random_list(length):
    return [random.randint(1, 1000) for _ in range(length)]

# Типы списков для тестирования
list_types = {
    "Отсортированный": generate_sorted_list,
    "Отсортированный в обратном порядке": generate_reverse_sorted_list,
    "Почти отсортированный": generate_almost_sorted_list,
    "Случайный": generate_random_list
}

# Алгоритмы для тестирования
sorting_algorithms = {
    "Пузырёк": bubble_sort,
    "Камешек-пузырёк": cocktail_shaker_sort
}

# Длины списков для тестирования
test_lengths = [10, 50, 100]

# Проведение тестирования
for list_type, generate_func in list_types.items():
    for alg_name, sort_algorithm in sorting_algorithms.items():
        print(f"Тип списка: {list_type}, Алгоритм сортировки: {alg_name}")
        for length in test_lengths:
            test_list = generate_func(length)
            sorted_list, comparisons, swaps = sort_algorithm(test_list.copy())
            print(f"Длина списка: {length}")
            print(f"Отсортированный список: {sorted_list}")
            print(f"Количество сравнений: {comparisons}")
            print(f"Количество перестановок: {swaps}")
            print("------------------------")
```

```python
# 15
def selection_sort_sublists(arr, sublist_size):
    n = len(arr)
    
    for i in range(0, n, sublist_size):
        sublist = arr[i:i+sublist_size]
        sublist.sort()  # Сортируем подсписок

        for j in range(i, min(i+sublist_size, n)):
            arr[j] = sublist[j - i]

    return arr

# Генерация случайного списка
def generate_random_list(length):
    return [random.randint(1, 100) for _ in range(length)]

# Тестирование модификации сортировки выбором на основе подсписков
def test_selection_sort_sublists(sublist_size, test_lengths):
    results = {}
    
    for length in test_lengths:
        comparisons = []
        swaps = []
        
        for _ in range(num_tests):
            test_list = generate_random_list(length)
            sorted_list = selection_sort_sublists(test_list.copy(), sublist_size)
            comparisons.append(0)  # Здесь просто обнуляем, так как для selection sort количество сравнений не всегда очевидно
            swaps.append(0)  # Аналогично для перестановок
        
        results[length] = {
            'min_comparisons': min(comparisons),
            'avg_comparisons': statistics.mean(comparisons),
            'max_comparisons': max(comparisons),
            'min_swaps': min(swaps),
            'avg_swaps': statistics.mean(swaps),
            'max_swaps': max(swaps)
        }
    
    return results

# Параметры тестирования
sublist_size = 5
test_lengths = [10, 50, 100]
num_tests = 5

# Тестирование модификации сортировки выбором на основе подсписков
results = test_selection_sort_sublists(sublist_size, test_lengths)

# Вывод результатов
for length, result in results.items():
    print(f"Длина списка: {length}")
    print(f"Минимальное количество сравнений: {result['min_comparisons']}")
    print(f"Среднее количество сравнений: {result['avg_comparisons']}")
    print(f"Максимальное количество сравнений: {result['max_comparisons']}")
    print(f"Минимальное количество перестановок: {result['min_swaps']}")
    print(f"Среднее количество перестановок: {result['avg_swaps']}")
    print(f"Максимальное количество перестановок: {result['max_swaps']}")
    print("---------------------")
```

```python
# 16
def partition(arr, low, high):
    pivot = arr[high]
    i = low - 1
    comparisons = 0
    swaps = 0

    for j in range(low, high):
        comparisons += 1
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
            swaps += 1

    arr[i + 1], arr[high] = arr[high], arr[i + 1]
    swaps += 1
    return i + 1, comparisons, swaps

def quick_sort(arr, low, high):
    if low < high:
        pi, comparisons, swaps = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

# Пример использования
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quick_sort(arr, 0, n - 1)
print("Отсортированный массив:", arr)

```

```python
# 17
import random

def partition(arr, low, high):
    pivot_index = random.randint(low, high)  # Выбираем случайный индекс в диапазоне [low, high]
    pivot = arr[pivot_index]
    arr[pivot_index], arr[high] = arr[high], arr[pivot_index]

    i = low - 1
    for j in range(low, high):
        if arr[j] < pivot:
            i += 1
            arr[i], arr[j] = arr[j], arr[i]
    arr[i+1], arr[high] = arr[high], arr[i+1]
    return i + 1

def quick_sort(arr, low, high):
    if low < high:
        pi = partition(arr, low, high)
        quick_sort(arr, low, pi - 1)
        quick_sort(arr, pi + 1, high)

# Пример использования
arr = [10, 7, 8, 9, 1, 5]
n = len(arr)
quick_sort(arr, 0, n-1)
print("Отсортированный массив:", arr)
```

```python
# 18
import heapq

def merge_sorted_lists(sorted_lists):
    merged_list = []
    heap = [(lst[0], i, 0) for i, lst in enumerate(sorted_lists) if lst]
    heapq.heapify(heap)
    
    while heap:
        val, list_index, element_index = heapq.heappop(heap)
        merged_list.append(val)
        if element_index + 1 < len(sorted_lists[list_index]):
            next_tuple = (sorted_lists[list_index][element_index + 1], list_index, element_index + 1)
            heapq.heappush(heap, next_tuple)
    
    return merged_list

# Пример нескольких отсортированных списков
sorted_lists = [
    [1, 3, 5],
    [2, 4, 6],
    [0, 7, 8]
]

result = merge_sorted_lists(sorted_lists)
print(result)
# 1. Создаем пустой список, в который будем добавлять элементы.
# 2. Берем по одному элементу из каждого отсортированного списка и сравниваем их.
# 3. Из выбранных элементов добавляем минимальный в новый список.
# 4. Повторяем шаги 2-3, пока все списки не будут полностью пройдены.
```

```python
# 19
def shell_sequence(n):
    sequence = []
    k = 1
    while (2**k - 1) < n:
        sequence.insert(0, 2**k - 1)
        k += 1
    return sequence

# Пример использования
n = 10
sequence = shell_sequence(n)
print(sequence)
```
